## Part-1 (for `row`)

```python
driver.switch_to_window(driver.window_handles[0])

all_license_det=[]
business_names=[]
address=[]
phone=[]

for i in range(1,len(names)+1):
    driver.find_element_by_xpath(xpaths[i-1]).click()
    driver.switch_to_window(driver.window_handles[1])
    page_source = driver.page_source
    soup =  BeautifulSoup(page_source, 'lxml')
    
    #getting headings
    headings = [i.text for i in soup.findAll('th')]
    #headings
    n=headings.index("Registration Status")
    #n
    
    headings=headings[:n+1]
    headings
    #len(headings)
    
    #getting actual data (row)
    trTag = soup.find_all("tr", {"class": "row"})
    data=[]
    for tag in trTag:
        tdTags = tag.find_all("td")
        for tag in tdTags:
            data.append(tag.text)
    data=data[4:]
    data
    n_data=[]
    for i in data:
        n_data.append(i.strip())
    #n_data
    #len(n_data)
    n_data=n_data[:-(len(n_data)-len(headings))]
    #n_data
    #len(n_data)
    all_license_det=all_license_det+ n_data
    
    #getting address
    trTag = soup.find_all("tr", {"class": "alternating-row"})
    data=[]
    for tag in trTag:
        tdTags = tag.find_all("td")
        for tag in tdTags:
            data.append(tag.text)
    add=data[1].strip()
    add=data[1].replace('\n',' ')
    address.append(add)

    
    
    
    
    
    
    #getting phone number
    trTag = soup.find_all("tr", {"class": "row"})
    data=[]
    for tag in trTag:
        tdTags = tag.find_all("td")
        for tag in tdTags:
            data.append(tag.text)
    ph=data[3]
    #ph
    phone.append(ph)
    
    
    trTag = soup.find_all("tr", {"class": "row"})
    data=[]
    for tag in trTag:
        tdTags = tag.find_all("td")
        for tag in tdTags:
            data.append(tag.text)
    business_name=data[1]
    business_names.append(business_name)
    
    
    
    
    time.sleep(1)
    driver.switch_to_window(driver.window_handles[0])
```
